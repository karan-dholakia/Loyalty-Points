# Calculating loyalty points

```python
#import libraries like pandas and numpy for data munipulation

import pandas as pd 
import numpy as np 
```

```python
user = pd.read_csv("user_gameplay.csv")       #loading user game played dataset
dep = pd.read_csv("deposite.csv")             #loading deposite dataset    
withdrawal = pd.read_csv("withdrawal.csv")    #lading withdrawal dataset
```

```python
#converting the datetime from object to datetime 

user['Datetime'] = pd.to_datetime(user['Datetime'])
dep['Datetime'] = pd.to_datetime(dep['Datetime'])
withdrawal['Datetime'] = pd.to_datetime(withdrawal['Datetime'])
```

```python
#divide the data set of user game played in Slot1 & Slot2 according to asked date  

S1_2oct_user = user[(user['Datetime'] >= '02/10/2022 00:00:00') & (user['Datetime'] <= '02/10/2022 12:00:00')]

S2_16oct_user = user[ (user['Datetime'] > '16-10-2022 12:00:00') & (user['Datetime'] < '17-10-2022 00:00:00')]

S1_18oct_user = user[ (user['Datetime'] > '18-10-2022 00:00:00') & (user['Datetime'] < '18-10-2022 12:00:00')]

S2_26oct_user =  user[ (user['Datetime'] > '26-10-2022 12:00:00') & (user['Datetime'] < '27-10-2022 00:00:00')]
 
```

```python
#divide the data set of user deposite in Slot1 & Slot2 according to asked date 

S1_2oct_dep = dep[ (dep['Datetime'] > '02-10-2022 00:00:00') & (dep['Datetime'] < '02-10-2022 12:00:00')]

S2_16oct_dep = dep[ (dep['Datetime'] > '16-10-2022 12:00:00') & (dep['Datetime'] < '17-10-2022  00:00:00')]

S1_18oct_dep = dep[ (dep['Datetime'] > '18-10-2022 00:00:00') & (dep['Datetime'] < '18-10-2022 12:00:00')]

S2_26oct_dep =  dep[ (dep['Datetime'] > '26-10-2022 12:00:00') & (dep['Datetime'] < '27/10/2022 00:00:00')]
 
```

```python
#divide the data set of user withdrawal in Slot1 & Slot2 according to asked date 

S1_2oct_withdrawal = withdrawal[ (withdrawal['Datetime'] > '02-10-2022 00:00:00') & (withdrawal['Datetime'] < '02-10-2022 12:00:00')]

S2_16oct_withdrawal = withdrawal[ (withdrawal['Datetime'] > '16-10-2022 12:00:00') & (withdrawal['Datetime'] < '17-10-2022  00:00:00')]

S1_18oct_withdrawal = withdrawal[ (withdrawal['Datetime'] > '18-10-2022 00:00:00') & (withdrawal['Datetime'] < '18-10-2022 12:00:00')]

S2_26oct_withdrawal =  withdrawal[ (withdrawal['Datetime'] > '26-10-2022 12:00:00') & (withdrawal['Datetime'] < '27/10/2022 00:00:00')]
 
```

```python
#finding loyalty points for Slot 1 2nd october 

sum_games_2 = S1_2oct_user.groupby(by = ['User ID'], as_index = False).agg({'Games Played': 'sum'}) #total games played by user in slot 1 

sum_dep_2 = S1_2oct_dep.groupby(by = ['User Id'], as_index = False).agg({'Amount':'sum'})

sum_withdraw_2 = S1_2oct_withdrawal.groupby(by = ['User Id'], as_index = False).agg({'Amount':'sum'})


Loyalty_points_2 = []

for i in S1_2oct_user['User ID']:
    total_games_played = sum_games_2[sum_games_2['User ID'] == i]['Games Played'].values 

    if i in S1_2oct_dep['User Id']:
        deposite = sum_dep_2[sum_dep_2['User Id']==i]['Amount'].values
    else:
        deposite = 0
    
    if i in S1_2oct_withdrawal['User Id']:
        withdraw = sum_withdraw_2[sum_withdraw_2['User Id']==i]['Amount'].values
    else:
        withdraw = 0
    
        
    Loyalty_points_2.append(((0.01*deposite) + (0.005*withdraw ) + (0.2*total_games_played)))
    
    
Loyalty_points_2 = pd.DataFrame(Loyalty_points_2, columns = ['Loyalty Points'], index = None)

Loyalty_points_S1 = Loyalty_points_2.assign(UserID = S1_2oct_user['User ID'].values)

print("Loyalty points in Slot 1 2nd October")
Loyalty_points_S1
```
